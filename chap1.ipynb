## 1. (15 分) $\Re\left(\frac{1}{z}\right)=2$ 在复平面上具有怎样的意义？

</font>

<font color=black size=4>
<font color=blue>解：</font>

令

$$z=x+\mathrm{i}y$$

则 

$$\frac{1}{z} = \frac{1}{x+\mathrm{i}y} = \frac{(x-\mathrm{i}y)}{(x+\mathrm{i}y)(x-\mathrm{i}y)} = \frac{x}{x^2+y^2} - \mathrm{i}\frac{y}{x^2+y^2}$$

$$\Re\left(\frac{1}{z}\right)  = \frac{x}{x^2+y^2} = 2$$
    
求解上述方程：

$$2x^2-x+2y^2 = 2\left(x^2-2\cdot\frac{1}{4}x+\frac{1}{16}\right) + 2y^2 = 2(x-\frac{1}{4})^2+2y^2 = \frac{1}{8}$$
    
整理可得，

$$\left(x-\frac{1}{4}\right)^2+y^2 = \left( \frac{1}{4} \right)^2$$

</font>

```python
fig, ax = plt.subplots(dpi=120)
cc = plt.Circle((0.25,0.0), 0.25, color='red', lw=2, fill=False, zorder=5)
ax.set_aspect(1)
ax.add_artist(cc)

ax.spines[["left", "bottom"]].set_position(("data", 0))
ax.spines[["top", "right"]].set_visible(False)
ax.plot(1, 0, ">k", transform=ax.get_yaxis_transform(), clip_on=False)
ax.plot(0, 1, "^k", transform=ax.get_xaxis_transform(), clip_on=False)

ax.set_xlim(-0.25,0.75)
ax.set_ylim(-0.50,0.50)
ax.set_xticks([-0.25,0.0, 0.25, 0.50, 0.75])
ax.set_yticks([-0.50,-0.25,0.0, 0.25, 0.50])
ax.axvline(0.0, color='k', lw=1.0, zorder=1)
ax.axhline(0.0, color='k', lw=1.0, zorder=1)
ax.scatter(0.25, 0.0, s=40.0, zorder=5)
plt.show()
```

## 2. (10分) 将复数$\frac{1-\mathrm{i}}{1+\mathrm{i}}$ 分别用代数式、三角式、指数式的形式表示出来。
    
</font>

<font color=black size=4> 
<font color=blue>解：</font>

把 $\displaystyle{\frac{1-\mathrm{i}}{1+\mathrm{i}}}$ 变成实部和虚部的形式：

$$
\begin{array}{rlr}
\displaystyle{\frac{1-\mathrm{i}}{1+\mathrm{i}}} & = \displaystyle{\frac{(1-\mathrm{i})(1-\mathrm{i})}{(1+\mathrm{i})(1-\mathrm{i})}} = \displaystyle{\frac{-2\mathrm{i}}{2}} &  \\
 & = -\mathrm{i} & (代数式)\\
 & = \displaystyle{\cos(\frac{3\pi}{2}+2n\pi) + \mathrm{i}\cdot\sin(\frac{3\pi}{2}+2n\pi)} & (三角式) \\
 & = \displaystyle{e^{\left[\mathrm{i}\cdot(\frac{3\pi}{2}+2n\pi)\right]}} & (指数式)
\end{array}
$$

</font>

## 4. (15分) 求解方程 $\sin(z) = 2$。

</font>

<font size=4 color=black>
<font color=blue>解：</font>

已知欧拉公式
$$
\begin{array}{rl}
\cos{x}+\mathrm{i}\sin{x} &= e^{\mathrm{i}x}  \\
\cos{x}-\mathrm{i}\sin{x} &= e^{-\mathrm{i}x}
\end{array}
$$
所以我们可以得到
$$\sin(z) = \frac{1}{2\mathrm{i}}\left( e^{\mathrm{i}z}-e^{-\mathrm{i}z} \right) = 2$$
令 $z=x+\mathrm{i}y$
则 
$$
\begin{array}{rl}
e^{\mathrm{i}z}-e^{-\mathrm{i}z} & = e^{-y+\mathrm{i}x} - e^{y-\mathrm{i}x} \\
  & = e^{-y}e^{\mathrm{i}x} - e^y e^{-\mathrm{i}x}  \\
  & = e^{-y}(\cos{x}+\mathrm{i}\sin{x}) -e^y(\cos{x}-\mathrm{i}\sin{x}) \\
  & = \left( e^{-y}\cos{x}-e^y\cos{x} \right) + \mathrm{i}\left( e^{-y}\sin{x} + e^y\sin{x} \right) \\
  & = 4\mathrm{i} 
\end{array}
$$
上式成立，则：
$$
\left\{ 
  \begin{array}{rl}
    e^{-y}\cos{x} - e^y\cos{x} &= 0  \\
    e^{-y}\sin{x} + e^y\sin{x} &= 4
  \end{array}
\right.
$$
该方程组成立有下列两种情况，

1. $\cos{x} \neq 0$
$$e^{-y} = e^{y}$$
方程组无解。

2. $\cos{x} = 0$
$$
x = (n+\frac{1}{2})\pi \Rightarrow \sin{x} = (-1)^n
$$
代入到原方程
$$
e^{-y}(-1)^n + e^y(-1)^n = 4 \\
e^{2y} - 4\cdot (-1)^n \cdot e^y + 1 = 0
$$
该方程的解为:
$$e^y=2\cdot(-1)^n \pm \sqrt{3}$$
n为奇数时，上式不成立；当n为偶数时，$y=\ln{ (2\pm\sqrt{3}) }$.
因此，
$$z=x+\mathrm{i}y = \left( 2k+\frac{1}{2} \right)\pi + \mathrm{i}\cdot\ln{ (2\pm\sqrt{3}) }$$

</font>

```python
fig, ax = plt.subplots(dpi=120)
ax.set_aspect(1)
ax.spines[["left", "bottom"]].set_position(("data", 0))
ax.spines[["top", "right"]].set_visible(False)
ax.plot(1, 0, ">k", transform=ax.get_yaxis_transform(), clip_on=False)
ax.plot(0, 1, "^k", transform=ax.get_xaxis_transform(), clip_on=False)
ax.set_xlim(-6.0, 10.0)
ax.set_ylim(-2.0,  2.0)
ax.axvline(0.0, color='k', lw=1.0, zorder=1)
ax.axhline(0.0, color='k', lw=1.0, zorder=1)
ax.scatter((-2+0.5)*np.pi, np.log(2.0+np.sqrt(3.0)), s=40.0, zorder=10, marker='^', color='red')
ax.scatter((-2+0.5)*np.pi, np.log(2.0-np.sqrt(3.0)), s=40.0, zorder=10, marker='^', color='blue')
ax.scatter( 0.5*np.pi, np.log(2.0+np.sqrt(3.0)), s=40.0, zorder=10, marker='o', color='red')
ax.scatter( 0.5*np.pi, np.log(2.0-np.sqrt(3.0)), s=40.0, zorder=10, marker='o', color='blue')
ax.scatter((2+0.5)*np.pi, np.log(2.0+np.sqrt(3.0)), s=40.0, zorder=10, marker='s', color='red')
ax.scatter((2+0.5)*np.pi, np.log(2.0-np.sqrt(3.0)), s=40.0, zorder=10, marker='s', color='blue')
ax.text(-5.0, 0.4, r"$k=-1$", size=16)
ax.text( 1.4, 0.4, r"$k= 0$", size=16)
ax.text( 6.5, 0.4, r"$k=+1$", size=16)
plt.show()
```

## 5. (15分) 已知解析函数 $f(z)$ 的实部 $u(x,y)=e^x\sin{y}$，求该解析函数。

</font>

<font color=red size=5>
<font color=blue>解：</font>

柯西-黎曼方程(Cauchy-Riemann Equations)：
$$
\left\{
\begin{array}{rl}
    \frac{\partial u}{\partial x} &= \frac{\partial v}{\partial y}, \\
    \frac{\partial v}{\partial x} &= -\frac{\partial u}{\partial y}.  
\end{array}
\right.
$$
</font>
<font color=black size=4>

$$
\begin{array}{rl}
\frac{\partial u}{\partial x} & = e^x\sin{y} = \frac{\partial v}{\partial y} \\
\frac{\partial u}{\partial y} & = e^x\cos{y} = -\frac{\partial v}{\partial x}
\end{array}
$$
把$v$写成全微分的形式，
$$\mathrm{d}v = -e^x\cos{y}\mathrm{d}x + e^x\sin{y}\mathrm{d}y = \mathrm{d} (-e^x\cos{y})$$
则，
$$v = -e^x\cos{y} + c$$
所以该解析函数为，
$$
\begin{array}{rl}
    f(z) &= e^x\sin{y} - \mathrm{i}e^x\cos{y}+\mathrm{i}c \\
         &= e^x(\sin{y} - \mathrm{i}\cos{y}) + \mathrm{i}c \\
         &= e^x(-\mathrm{i})e^{\mathrm{i}y} + \mathrm{i}c \\
         &= (-i) e^{x+\mathrm{i}y} + \mathrm{i}c = (-\mathrm{i})e^z + \mathrm{i}c
\end{array}
$$

</font>

```python
def func(x,y,c):
    return -1*1j*np.exp(x+1j*y)+1j*c

for cc in [-1.0, 0.0, 1.0]:

    print("c=",cc)
    
    xx = np.linspace(-1.0, 1.0, num=100)
    yy = np.linspace(-1.0, 1.0, num=100)
    zz = np.array([ func(ix,iy,cc) for ix in xx for iy in yy ]).reshape(100, 100)

    xx, yy = np.meshgrid(xx, yy)

    fig, ax = plt.subplots(nrows=1,ncols=3,subplot_kw={"projection": "3d"}, figsize=(18,6), dpi=120, sharex=True, sharey=True)
    ax[0].plot_surface(xx, yy, zz.real, cmap='coolwarm', edgecolor='none')
    ax[0].set_xlabel(r"$x$", size=20)
    ax[0].set_ylabel(r"$y$", size=20)
    ax[0].set_zlabel(r"$\Re{f(z)}$", size=20)
    ax[0].set_xticks([-1.0,-0.5, 0.0, 0.5, 1.0])
    ax[0].set_yticks([-1.0,-0.5, 0.0, 0.5, 1.0])
    ax[0].view_init(20, 40)

    ax[1].plot_surface(xx, yy, zz.imag, cmap='coolwarm', edgecolor='none')
    ax[1].set_xlabel(r"$x$", size=20)
    ax[1].set_ylabel(r"$y$", size=20)
    ax[1].set_zlabel(r"$\Im{f(z)}$", size=20)
    ax[1].view_init(20, 40)

    ax[2].plot_surface(xx, yy, np.absolute(zz), cmap='coolwarm', edgecolor='none')
    ax[2].set_xlabel(r"$x$", size=20)
    ax[2].set_ylabel(r"$y$", size=20)
    ax[2].set_zlabel(r"$|f(z)|$", size=20)
    ax[2].view_init(20, 30)
    plt.show()
```

## 6. (15分）已知解析函数 $f(z)$ 的虚部$v(x,y) = \frac{y}{x^2+y^2}$，且$f(2)=0$，求该解析函数。 

</font>

<font size=4 color=black>
<font color=blue>解：</font>

由柯西-黎曼条件：
$$
\begin{array}{rl}
    \displaystyle{\frac{\partial v}{\partial x}} & = \displaystyle{-\frac{2xy}{(x^2+y^2)^2}} = -\frac{\partial u}{\partial y} \\
    \displaystyle{\frac{\partial v}{\partial y}} & = \displaystyle{\frac{x^2-y^2}{(x^2+y^2)^2}} = \frac{\partial u}{\partial x} \\
\end{array}
$$

$$
\begin{array}{rl}
u(x,y) &= \displaystyle{\int{\frac{x^2-y^2}{(x^2+y^2)^2}\cdot\mathrm{d}x} + \int{\frac{2xy}{(x^2+y^2)^2}\cdot\mathrm{d}y} + c} \\
       &= \displaystyle{\int{\mathrm{d}\left( \frac{-x}{x^2+y^2} \right)} + c} \\
       &= \displaystyle{-\frac{x}{x^2+y^2} + c}
\end{array}
$$
因此，
$$
\begin{array}{rl}
    f(z) &= u(x,y) + \mathrm{i}v(x,y) \\
         &= \displaystyle{-\frac{x}{x^2+y^2} + c + \mathrm{i}\frac{y}{x^2+y^2}} \\
         &= \displaystyle{-\frac{x-\mathrm{i}y}{x^2+y^2}+c = -\frac{x-\mathrm{i}y}{(x+\mathrm{i}y)(x-\mathrm{i}y)}+c} \\
         &= \displaystyle{-\frac{1}{z} + c}
\end{array}
$$
已知$f(2)=0$，可得$c=\frac{1}{2}$，

所以该解析函数为：
$$
f(z) = -\frac{x}{x^2+y^2} +\frac{1}{2} + \mathrm{i}\frac{y}{x^2+y^2} = -\frac{1}{z} + \frac{1}{2}
$$


</font>

```python
def func(x, y):
    return -x/(x**2.0+y**2.0)+0.5+1j*(y/(x**2.0+y**2.0))

xx = np.linspace(-0.01, 0.01, num=100)
yy = np.linspace(-0.01, 0.01, num=100)
zz = np.array([ func(ix,iy) for ix in xx for iy in yy ]).reshape(100, 100)

xx, yy = np.meshgrid(xx, yy)

fig, ax = plt.subplots(nrows=1,ncols=3,subplot_kw={"projection": "3d"}, figsize=(18,6), dpi=120, sharex=True, sharey=True)
ax[0].plot_surface(xx, yy, zz.real, cmap='coolwarm', edgecolor='none')
ax[0].set_xlabel(r"$x$", size=20)
ax[0].set_ylabel(r"$y$", size=20)
ax[0].set_zlabel(r"$\Re{f(z)}$", size=20)
ax[0].set_zlim(-2000, 2000)
ax[0].set_xticks([-0.01, 0.0, 0.01])
ax[0].set_yticks([-0.01, 0.0, 0.01])
ax[0].set_zticks([-2000, 0, 1000])
ax[0].view_init(20, 40)

ax[1].plot_surface(xx, yy, zz.imag, cmap='coolwarm', edgecolor='none')
ax[1].set_xlabel(r"$x$", size=20)
ax[1].set_ylabel(r"$y$", size=20)
ax[1].set_zlabel(r"$\Im{f(z)}$", size=20)
ax[1].set_zlim(-2000, 2000)
ax[1].set_zticks([-2000, 0, 2000])
ax[1].view_init(20, 40)

ax[2].plot_surface(xx, yy, np.absolute(zz), cmap='coolwarm', edgecolor='none')
ax[2].set_zlabel(r"$|f(z)|$", size=20)
ax[2].set_zlim(0, 4000)
ax[2].set_zticks([0, 2000, 4000])
ax[2].view_init(20, 40)
plt.show()
```

<font color=blue>考虑用极坐标系来求解</font>

极坐标系的柯西-黎曼条件为：
$$
\begin{align*}
    \frac{\partial u}{\partial \rho} & = \frac{1}{\rho} \cdot \frac{\partial v}{\partial \psi} \\
    \frac{1}{\rho} \cdot \frac{\partial u}{\partial \psi} & = -\frac{\partial v}{\partial \rho}
\end{align*}
$$

<font color=black size=4>

令 $x=\rho\cos{\psi}$, $y=\rho\sin{\psi}$，则原解析函数的虚部可以写为：
$$
v(x,y) = \frac{y}{x^2+y^2} = \displaystyle{\frac{\rho\sin{\psi}}{\rho^2\cos^2{\psi} + \rho^2\sin^2{\psi}} = \frac{1}{\rho}\cdot\sin{\psi}} = v(\rho,\psi)
$$
由柯西-黎曼条件得：
$$
\left\{
\begin{array}{rl}
    \displaystyle{\frac{\partial v}{\partial \psi}} & = \displaystyle{\rho\cdot\frac{\partial u}{\partial \rho}} = \displaystyle{\frac{1}{\rho}\cdot\cos{\psi}} \\
    \displaystyle{\frac{\partial v}{\partial \rho}} & = \displaystyle{-\frac{1}{\rho}\cdot \frac{\partial u}{\partial \psi}} = \displaystyle{-\frac{1}{\rho^2}\cdot\sin{\psi}}
\end{array}
\right.
$$
因此，
$$
\frac{\partial u}{\partial \rho} = \displaystyle{\frac{1}{\rho^2}\cdot\cos{\psi}}, \quad \frac{\partial u}{\partial \psi} = \frac{1}{\rho}\cdot\sin{\psi}
$$

写成全微分的形式，
$$
\begin{array}{rl}
\mathrm{d}u &= \displaystyle{\frac{1}{\rho^2}\cos{\psi}\cdot\mathrm{d}\rho + \frac{1}{\rho}\sin{\psi}\cdot\mathrm{d} \psi} \\
            &= \displaystyle{\mathrm{d}\left( -\frac{1}{\rho}\cos{\psi} \right)}
\end{array}
$$
所以原解析函数的实部为，
$$
u(\rho,\psi) = -\frac{1}{\rho}\cos{\psi} + c
$$
原解析函数为，
$$
\begin{array}{rl}
f(z) &= \displaystyle{-\frac{1}{\rho}\cos{\psi} + c + \mathrm{i}\cdot\frac{1}{\rho}\sin{\psi}} \\
     &= \displaystyle{-\frac{1}{\rho}(\cos{\psi}-\mathrm{i}\sin{\psi}) + c} = -\frac{1}{z} + c
\end{array}
$$
由于，
$$
z=2 \Rightarrow \rho=2, \psi=0
$$
可得
$$
f(2) = -\frac{1}{2}+c+\mathrm{i}\cdot\frac{1}{2}\cdot 0 =  0 \Rightarrow c=\frac{1}{2}
$$
所以该解析函数在极坐标系下可以表示为，
$$
f(z) = -\frac{1}{\rho}\cos{\psi} + \frac{1}{2} + \mathrm{i}\cdot\frac{1}{\rho}\sin{\psi} = -\frac{1}{z}+\frac{1}{2}
$$

</font>
