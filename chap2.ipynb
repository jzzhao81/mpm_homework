## 2. (25分) 按给定的路径计算积分$\displaystyle{\int^1_{-1}{\frac{\mathrm{d}z}{z}}}$：
    
(1) 沿路径$ C_1 $: $ |z|=1 $的上半圆周；
    
(2) 沿路径$ C_2 $: $ |z|=1 $的下半圆周。
    
</font>

```python
fig, ax = plt.subplots(dpi=120)
ax.set_aspect(1)
circle1 = Wedge((0,0), 1,   0, 180, ec='red',  fc='None', lw=2, fill='None', zorder=1)
circle2 = Wedge((0,0), 1, 180, 360, ec='blue', fc='None', lw=2, fill='None', zorder=1)
ax.add_artist(circle1)
ax.add_artist(circle2)
ax.axvline(0, color='k', lw=2, zorder=5)
ax.axhline(0, color='k', lw=2, zorder=5)
ax.spines[["left", "bottom"]].set_position(("data", 0))
ax.spines[["top", "right"]].set_visible(False)
ax.plot(1, 0, ">k", transform=ax.get_yaxis_transform(), clip_on=False)
ax.plot(0, 1, "^k", transform=ax.get_xaxis_transform(), clip_on=False)
ax.set_xlim(-1.5, 1.5)
ax.set_ylim(-1.5, 1.5)
ax.set_xticks([-1,0,1])
ax.set_xticks([-1.5,-0.5, 0.5, 1.5],minor=True)
ax.set_yticks([-1,0,1])
ax.set_yticks([-1.5,-0.5, 0.5, 1.5],minor=True)
plt.show()
```

<font color=blue>解：</font>
<font color=black size=4>

令 $z=e^{\mathrm{i}\phi}$，
    
则 $\mathrm{d}z=\mathrm{i}e^{\mathrm{i}\phi}\mathrm{d}\phi$

(1) 沿路径<font color=red>$C_1$: $|z|=1$的上半圆周</font>积分:
$$
\int^1_{-1}\frac{\mathrm{d}z}{z} = \int^0_{\pi}\frac{\mathrm{i}e^{\mathrm{i}\phi}\mathrm{d}\phi}{e^{\mathrm{i}\phi}} = \int^0_{\pi} \mathrm{i}\mathrm{d}\phi = -\mathrm{i}\pi
$$

(2) 沿路径<font color=blue>$C_1$: $|z|=1$的下半圆周</font>积分:
$$
\int^1_{-1}\frac{\mathrm{d}z}{z} = \int^{2\pi}_{\pi}\frac{\mathrm{i}e^{\mathrm{i}\phi}\mathrm{d}\phi}{e^{\mathrm{i}\phi}} = \int^{2\pi}_{\pi} \mathrm{i}\mathrm{d}\phi = \mathrm{i}\pi
$$

</font>

## 3. (25分）计算积分$\displaystyle{\oint_{C}{\frac{\sin{\frac{\pi z}{4}}}{z^2-1}\mathrm{d}z}}$，$C$分别为：
    
(1) $|z|=\frac{1}{2}$； (2) $|z-1|=1$； (3) $|z|=3$
    
</font>

```python
fig, ax = plt.subplots(dpi=120)
cc1 = plt.Circle((0.0,0.0), 0.5, color='red',   lw=2, fill=False, zorder=5)
cc2 = plt.Circle((1.0,0.0), 1.0, color='blue',  lw=2, fill=False, zorder=5)
cc3 = plt.Circle((0.0,0.0), 3.0, color='green', lw=2, fill=False, zorder=5)
ax.set_aspect(1)
ax.add_artist(cc1)
ax.add_artist(cc2)
ax.add_artist(cc3)
ax.spines[["left", "bottom"]].set_position(("data", 0))
ax.spines[["top", "right"]].set_visible(False)
ax.plot(1, 0, ">k", transform=ax.get_yaxis_transform(), clip_on=False)
ax.plot(0, 1, "^k", transform=ax.get_xaxis_transform(), clip_on=False)
ax.set_xlim(-3.5, 3.5)
ax.set_ylim(-3.5, 3.5)
ax.set_xticks([-3.0,-2.0,-1.0, 0.0, 1.0, 2.0, 3.0])
ax.set_yticks([-3.0,-2.0,-1.0, 0.0, 1.0, 2.0, 3.0])
ax.axvline(0.0, color='k', lw=1.0, zorder=1)
ax.axhline(0.0, color='k', lw=1.0, zorder=1)
ax.scatter( 1.0, 0.0, s=40.0, color='purple', zorder=5)
ax.scatter(-1.0, 0.0, s=40.0, color='purple', zorder=5)
plt.show()
```

<font color=blue>解：</font>

<font color=black size=4>
(1) 函数
    
$$f(z)=\displaystyle{\frac{\sin{\frac{\pi z}{4}}}{z^2-1}}$$
    
在单连通区域<font color=red>$\displaystyle{|z|=\frac{1}{2}}$</font>内解析，则根据<font color=red size=4>单连通区域柯西定理</font>
    
$$
\oint_l{f(z)}\mathrm{d}z = 0
$$
    
可知
    
$$
\oint_{|z|=\frac{1}{2}}{\frac{\sin{\frac{\pi z}{4}}}{z^2-1}\mathrm{d}z} = 0
$$

(2) 积分路径包含一个奇点<font color=blue>$z=1$</font>，根据柯西积分公式，
    
$$
\oint_{|z-1|=1}\frac{\sin{\frac{\pi z}{4}}}{z^2-1} \mathrm{d}z = \oint_{|z-1|=1}\frac{\frac{\sin{\frac{\pi z}{4}}}{z+1}}{z-1} \mathrm{d}z = 2\pi\mathrm{i}\times\frac{\sin{\frac{\pi z}{4}}}{z+1} \Bigg|_{z=1} = \frac{\sqrt{2}}{2}\pi\mathrm{i}
$$
 
同理$C$为$|z+1|=1$时，
    
$$
\oint_{|z+1|=1}\frac{\sin{\frac{\pi z}{4}}}{z^2-1} \mathrm{d}z = \oint_{|z+1|=1}\frac{\frac{\sin{\frac{\pi z}{4}}}{z-1}}{z+1} \mathrm{d}z = 2\pi\mathrm{i}\times\frac{\sin{\frac{\pi z}{4}}}{z-1} \Bigg|_{z=-1} = \frac{\sqrt{2}}{2}\pi\mathrm{i}
$$
    
(3) 积分路径包含两个奇点<font color=green>$z=\pm 1$</font>，原积分可以利用复连通区域柯西定理,
    
$$
\begin{array}{rl}
\displaystyle{\oint_{|z|=3}{\frac{\sin{\frac{\pi z}{4}}}{z^2-1}\mathrm{d}z}} &= \displaystyle{\oint_{|z-1|=1}\frac{\sin{\frac{\pi z}{4}}}{z^2-1} \mathrm{d}z + \oint_{|z+1|=1}\frac{\sin{\frac{\pi z}{4}}}{z^2-1} \mathrm{d}z} \\
& = \displaystyle{\frac{\sqrt{2}}{2}\pi\mathrm{i} + \frac{\sqrt{2}}{2}\pi\mathrm{i}} = \displaystyle{\sqrt{2}\pi\mathrm{i}}
\end{array}
$$
    
</font>

## 4. (25分) 利用柯西公式计算回路积分$\displaystyle{\oint_{|z|=1}{\frac{e^z}{z^3}\mathrm{d}z}}$
    
</font>

<font color=black size=4>
本题利用柯西积分公式的推论, <u>解析函数可以求导任意多次</u>：
    
$$
f^{(n)}(z) = \frac{n!}{2\pi\mathrm{i}}\oint_l\frac{f(\zeta)}{(\zeta-z)^{n+1}}\mathrm{d}\zeta
$$
    
在本题中，令$n=2, z=0$，则

$$
f^{(2)}(0) = \frac{2!}{2\pi\mathrm{i}}\oint_{l}\frac{f(\zeta)}{\zeta^3}\mathrm{d}\zeta
$$

所以，我们可以看出，计算本题中的积分可以通过:

$$
\oint_{|z|=1}\frac{e^z}{z^3}\mathrm{d}z = \oint_{|\zeta|=1}\frac{e^\zeta}{z^3}\mathrm{d}\zeta = \frac{2\pi\mathrm{i}}{2!}\cdot(e^\zeta)''\Bigg|_{\zeta=0} = \pi\mathrm{i}
$$

</font>
