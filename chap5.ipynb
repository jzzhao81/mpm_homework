## 1. (20分) 将周期函数$f(x)$展开为傅里叶级数。在$(-\pi,\pi)$这个周期上，函数可表示为：
   
$$
f(x)=
\begin{cases}
-1  \quad (-\pi< x< 0) \\
1   \quad (0\leq x \leq \pi)
\end{cases}
$$

</font>

```python
fig, ax = plt.subplots(dpi=150)

ax.set_aspect(1)

ax.spines[["left", "bottom"]].set_position(("data", 0))
ax.spines[["top", "right"]].set_visible(False)
ax.plot(1, 0, ">k", transform=ax.get_yaxis_transform(), clip_on=False)
ax.plot(0, 1, "^k", transform=ax.get_xaxis_transform(), clip_on=False)

ax.plot((-2*np.pi,-np.pi),(1,1), 'r-', lw=2)
ax.plot((-np.pi, 0),(-1,-1), 'r-', lw=2)
ax.plot((0, np.pi),( 1, 1), 'r-', lw=2)
ax.plot((np.pi, 2*np.pi),(-1,-1), 'r-', lw=2)

ax.plot((-np.pi,-np.pi),(-1, 1), 'r--', lw=2)
ax.plot(( np.pi, np.pi),(-1, 1), 'r--', lw=2)
ax.plot(( 0, 0),(-1, 1), 'r--', lw=2)

ax.set_xlim(-2*np.pi-0.8, 2*np.pi+0.8)
ax.set_ylim(-1.5, 1.5)
ax.set_xticks([-2*np.pi,-1*np.pi,1*np.pi,2*np.pi])
ax.set_xticklabels([r'$-2\pi$',r'$-\pi$',r'$\pi$',r'$2\pi$'])
ax.set_yticks([-1, 0, 1])
ax.set_yticklabels([r'-1',r'0', r'1'])

plt.show()
```

<font color=blue size=4>解：</font>
<font color=black size=4>
$$
\begin{align*}
f(x) &= a_0 + \sum^\infty_{k=1}{\left( a_k\cos{\frac{k\pi x}{l}}+b_k\sin{\frac{k\pi x}{l}} \right)} \\
a_0 &= \frac{1}{2\pi}\int^\pi_{-\pi}{f(\xi)\,\mathrm{d}\,\xi} = 0 \\
a_k &= \frac{1}{\pi}\int^\pi_{-\pi}{f(\xi)\cos{k\xi}\,\mathrm{d}\,\xi} = 0 \\
b_k &= \frac{1}{\pi}\int^\pi_{-\pi}{f(\xi)\sin{k\xi}\,\mathrm{d}\,\xi} = \frac{2}{\pi}\int^\pi_0{\sin{k\xi}\,\mathrm{d}\,\xi} = \frac{2}{k\pi}[1-(-1)^k]
\end{align*}
$$
代入系数可得，
$$
f(x) = \sum^\infty_{k=1}{b_k\sin{kx}} = \sum^\infty_{k=1}{\frac{2}{k\pi}[1-(-1)^k]\sin{kx}} = \sum^\infty_{n=0}{\frac{4}{(2n+1)\pi}\sin{(2n+1)x}}
$$
</font>

```python
def func51(n,x):
    rslt = np.zeros(x.shape[0])
    for ik in range(n):
        rslt += 4*np.sin((2*ik+1)*x)/(2*ik+1)/np.pi
    return rslt

fig, ax = plt.subplots(dpi=150)

ax.set_aspect(2)

ax.spines[["left", "bottom"]].set_position(("data", 0))
ax.spines[["top", "right"]].set_visible(False)
ax.plot(1, 0, ">k", transform=ax.get_yaxis_transform(), clip_on=False)
ax.plot(0, 1, "^k", transform=ax.get_xaxis_transform(), clip_on=False)

ax.plot((-2*np.pi,-np.pi),(1,1), 'r-', lw=2)
ax.plot((-np.pi, 0),(-1,-1), 'r-', lw=2)
ax.plot((0, np.pi),( 1, 1), 'r-', lw=2)
ax.plot((np.pi, 2*np.pi),(-1,-1), 'r-', lw=2)

ax.plot((-np.pi,-np.pi),(-1, 1), 'r--', lw=2)
ax.plot(( np.pi, np.pi),(-1, 1), 'r--', lw=2)
ax.plot(( 0, 0),(-1, 1), 'r--', lw=2)

xa = np.linspace(-7, 7, 501)
ya = func51(10, xa)

ax.plot(xa, ya, 'b-', lw=2)

ax.set_xlim(-2*np.pi-0.8, 2*np.pi+0.8)
ax.set_ylim(-1.5, 1.5)
ax.set_xticks([-2*np.pi,-1*np.pi,1*np.pi,2*np.pi])
ax.set_xticklabels([r'$-2\pi$',r'$-\pi$',r'$\pi$',r'$2\pi$'])
ax.set_yticks([-1, 0, 1])
ax.set_yticklabels([r'-1',r'0', r'1'])

plt.show()
```

## 2. (20分) 已知周期函数$f(x)=\displaystyle{\frac{u_0}{l}x} \quad (-l<x<l)$，，将其展开为傅里叶级数。

</font>

```python
fig, ax = plt.subplots(dpi=150)

ax.set_aspect(1)

ax.spines[["left", "bottom"]].set_position(("data", 0))
ax.spines[["top", "right"]].set_visible(False)
ax.plot(1, 0, ">k", transform=ax.get_yaxis_transform(), clip_on=False)
ax.plot(0, 1, "^k", transform=ax.get_xaxis_transform(), clip_on=False)

x1 = np.linspace(-3*np.pi, -np.pi, 101)
x2 = np.linspace(-np.pi, np.pi, 101)
x3 = np.linspace(np.pi, 3*np.pi, 101)
yy = np.linspace(-np.pi, np.pi, 101)

ax.plot(x1, yy, 'r-', lw=2)
ax.plot(x2, yy, 'r-', lw=2)
ax.plot(x3, yy, 'r-', lw=2)

ax.plot(( np.pi, np.pi), (-np.pi, np.pi), ls='--', color='red', lw=2)
ax.plot((-np.pi,-np.pi), (-np.pi, np.pi), ls='--', color='red', lw=2)

ax.set_xlim(-3*np.pi-0.8, 3*np.pi+0.8)
ax.set_ylim(-np.pi-0.5, np.pi+0.5)
ax.set_xticks([-3*np.pi,-2*np.pi,-1*np.pi,1*np.pi,2*np.pi,3*np.pi])
ax.set_xticklabels([r'$-3l$',r'$-2l$',r'$-l$',r'$l$',r'$2l$',r'$3l$'])
ax.set_yticks([-np.pi, np.pi])
ax.set_yticklabels([r'$-u_0$',r'$u_0$'])

plt.show()
```

<font color=blue size=4>解：</font>

<font color=black size=4>

已知该函数为奇函数，因此$a_0=a_k=0$
$$
\begin{align*}
    b_k &= \frac{2}{l}\int^l_0{f(\xi)\sin{\frac{k\pi\xi}{l}\mathrm{d}\xi}} = \frac{2}{l}\int^l_0{\frac{u_0}{l}\xi\sin{\frac{k\pi\xi}{l}\mathrm{d}\xi}} = \frac{(-1)^{k+1}2u_0}{k\pi} \\
    f(x)&=\sum^\infty_{k=1}{b_k\,\sin{\frac{k\pi x}{l}}} = \sum^\infty_{k=1}{\frac{(-1)^{k+1}2u_0}{k\pi}\sin{\frac{k\pi x}{l}}} \\
\end{align*}
$$

</font>

```python
def func52(k,x):
    rslt = np.zeros(x.shape[0])
    for ik in range(1,k):
        rslt += (-1.0)**(ik+1)*2.0*np.sin(ik*x)/ik
    return rslt

fig, ax = plt.subplots(dpi=150)

ax.set_aspect(1)

ax.spines[["left", "bottom"]].set_position(("data", 0))
ax.spines[["top", "right"]].set_visible(False)
ax.plot(1, 0, ">k", transform=ax.get_yaxis_transform(), clip_on=False)
ax.plot(0, 1, "^k", transform=ax.get_xaxis_transform(), clip_on=False)

x1 = np.linspace(-3*np.pi, -np.pi, 101)
x2 = np.linspace(-np.pi, np.pi, 101)
x3 = np.linspace(np.pi, 3*np.pi, 101)
yy = np.linspace(-np.pi, np.pi, 101)

ax.plot(x1, yy, 'r-', lw=2)
ax.plot(x2, yy, 'r-', lw=2)
ax.plot(x3, yy, 'r-', lw=2)

ax.plot(( np.pi, np.pi), (-np.pi, np.pi), ls='--', color='red', lw=2)
ax.plot((-np.pi,-np.pi), (-np.pi, np.pi), ls='--', color='red', lw=2)

xa = np.linspace(-3*np.pi, 3*np.pi, 501)
ya = func52(10, xa)

ax.plot(xa, ya, 'b-', lw=2)

ax.set_xlim(-3*np.pi-0.8, 3*np.pi+0.8)
ax.set_ylim(-np.pi-0.5, np.pi+0.5)
ax.set_xticks([-3*np.pi,-2*np.pi,-1*np.pi,1*np.pi,2*np.pi,3*np.pi])
ax.set_xticklabels([r'$-3l$',r'$-2l$',r'$-l$',r'$l$',r'$2l$',r'$3l$'])
ax.set_yticks([-np.pi, np.pi])
ax.set_yticklabels([r'$-u_0$',r'$u_0$'])

plt.show()
```

## 3. (20分) 非周期函数
$$
f(x)=
\begin{cases}
0 & (x<-\pi) \\
h & (-\pi\leq x\leq \pi) \\
0 & ( x>\pi)
\end{cases}
$$
展开为傅里叶积分。
    
</font>

```python
u0 = 1; ll = 2; aa = 1

fig, ax = plt.subplots(dpi=150)

ax.set_aspect(2)

ax.spines[["left", "bottom"]].set_position(("data", 0))
ax.spines[["top", "right"]].set_visible(False)
ax.plot(1, 0, ">k", transform=ax.get_yaxis_transform(), clip_on=False)
ax.plot(0, 1, "^k", transform=ax.get_xaxis_transform(), clip_on=False)

ax.plot((-aa, aa), (u0,u0), 'r-', lw=2)
ax.plot((-aa,-aa), ( 0,u0), 'r-', lw=2)
ax.plot(( aa, aa), ( 0,u0), 'r-', lw=2)
ax.plot((-7, -aa), ( 0, 0), 'r-', lw=2)
ax.plot(( aa,  7), ( 0, 0), 'r-', lw=2)

ax.set_xlim(-4, 4)
ax.set_ylim(-0.3, u0+0.3)
ax.set_xticks([-aa, aa])
ax.set_xticklabels([r'$-\pi$', r'$\pi$'])
ax.set_yticks([1])
ax.set_yticklabels([r'$h$'])

plt.show()
```

<font color=blue size=4>解：</font>

<font color=black size=4>

函数为偶函数
$$
\begin{align*}
A(\omega) &= \frac{2}{\pi}\int^\infty_0{f(\xi)\cos{\omega\xi}\,\mathrm{d}\xi} = \frac{2}{\pi}\int^\pi_0{h\cos{\omega\xi}\,\mathrm{d}\omega} = \frac{2h}{\pi\omega}\sin{\omega\pi} \\
f(x) &= \int^\infty_0{ \frac{2h}{\pi\omega} \sin{\pi\omega}\cdot\cos{\omega x}\,\mathrm{d}\omega }
\end{align*}
$$

</font>

```python
def func53(u0, l, a, k, x):
    return (2*u0*np.sin(k*np.pi*a/l)/(k*np.pi))*np.cos(k*np.pi*x/l)

def func54(u0, l, a, k, x):
    rslt = np.zeros(x.shape[0]) + (a*u0/l)
    for ik in range(1,k):
        rslt += func53(u0, l, a, ik, x)
    return rslt

u0 = 1; ll = 4; aa = 1

fig, ax = plt.subplots(dpi=150)

ax.set_aspect(2)

ax.spines[["left", "bottom"]].set_position(("data", 0))
ax.spines[["top", "right"]].set_visible(False)
ax.plot(1, 0, ">k", transform=ax.get_yaxis_transform(), clip_on=False)
ax.plot(0, 1, "^k", transform=ax.get_xaxis_transform(), clip_on=False)

ax.plot((-aa, aa), (u0,u0), 'r-', lw=2)
ax.plot((-aa,-aa), ( 0,u0), 'r-', lw=2)
ax.plot(( aa, aa), ( 0,u0), 'r-', lw=2)

xx = np.linspace(-ll, ll, 201)
y1 = func54(u0, ll, aa, 20, xx)

ax.plot(xx, y1, 'b-', alpha=0.9, lw=2)

ax.set_xlim(-ll-ll/5, ll+ll/5)
ax.set_ylim(-u0-0.3, u0+0.3)
ax.set_xticks([-aa, aa])
ax.set_xticklabels([r'$-\pi$', r'$\pi$'])
ax.set_yticks([1])
ax.set_yticklabels([r'h'])

plt.show()
```
